# Handling missing values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['license', 'house_rules'], errors='ignore')
```

```python
df.head()
```

```python
df['price'] = df['price'].replace('[\$,]', '',regex = True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex = True).astype(float)
```

```python
df.head()
```

# Remove duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

# Visualization

What is the distribution of listing prices?

```python
#df.head()
```

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'], bins = 50, kde = True, color = 'hotpink')
plt.title('Distribution of listing price')
plt.xlabel('price $')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows a fairly even distribution of listing prices across 
different price ranges, indicating no particular concentration of listings in any specific price range, The KDE line helps visulize this even spread more clearly confirming that the dataset contains listings with a wide variety of prices.

# What is the relationship between price and room type

```python
plt.figure(figsize= (10,6))
sns.boxplot(x = 'room type', y = 'price', hue = 'room type', data = df, palette='Set1')
plt.title('Price vs Room Type')
plt.xlabel('Room type')
plt.ylabel('Price ($)')
plt.show()
```

# How has the number of reviews change over the time

__Step 1__: df['last review'].dt.to_period('M')
Converts each date in 'last review' to a Year-Month period like:
'2020-01' (January 2020)
'2021-05' (May 2021)

__Step 2: .groupby(...)__
Groups all rows that have the same month and year.

__Step 3: .size()__
Counts how many times each Year-Month group appears.

```python
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize = (12,6))
reviews_over_time.plot(kind = 'line', color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel("Date")
plt.ylabel("Numbers of Reviews")
plt.show()      
```
